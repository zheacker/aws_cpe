## EC2 instance types
- general purpose
    - pretty balanced, good for web servers or code repositories
- compute optimized
    - ideal for compute intensive tasks like gaming servers, high-performance computing (HPC), scientific modeling
- memory optimized
    - good for memory intensive tasks
- accelerated computing
    - floating point number calculations, graphics processing, data pattern matching, they utilize hardware accelerators
- storage optimized
    - high performance for locally stored data

## Pricing
- on demand - pay per second or hour (depending on specifics)
- savings plan - low prices in exchange for a commitment of usage on 1 to 3 year contracts.
- reserved instances - good for predictable usage. You qualify once you commit to 1 or 3 year terms, and you can pay all up front, partial up front, or no up front.
- spot instances - cheaper than on demand, but aws can reclaim them whenever. You can spin back up later, so this can work for interruptable workloads, like batch workloads
- dedicated hosts - necessary for certain compliance requirements, because they're single tenancy

## Scaling EC2
Your IT resources are a classic utilization dilemma. In order to meet all of the demand, you *must* have a < 100% utilization rate. The lower your utilization, the better you can weather an extreme spike in demand.

So you can scale your EC2 instances to adjust to demand, keeping your utilization high enough to justify the cost, but infinitely variable, so that you can always spin up and meet new demand.

### Up vs. Out
Scaling up means adding power to the instance you're talking about. Scaling out means adding more instances to field more requests.

## Elastic Load Balancing
Elastic load balancing directs traffic amongst your EC2 instances, like a host at the door of the coffee shop. This routes new requests to different EC2 instances to ensure efficient handling. This is called load balancing, and you can run your own load balancer on AWS, or you can use Elastic Load Balancing (ELB) that they provide. It's a regional construct, so it meets high availability (HA) concern out of the box.

It's also not only for external traffic. It can sit between the front end and the back end as well, which means it can route effectively while the backend scales out and in. ELB is only 1 method of handling backend traffic, we'll dig into more later.

## Messaging & Queueing
In our coffee analogy, the cashier hands off the customer's order to the barista (sending a message). But if the barista is backed up or currently busy, we don't want to lose the message. If the front end and backend get out of sync, we run into problems. So we add a buffer, or queue into the process to handle the out of sync-ness.

A system is a "tightly coupled architecture" if there's no buffer, and applications communicate directly. "Loosely coupled architectures" are those where a single failure won't cascade through the system.

Amazon has 2 services that can help with things like this: Amazon Simple Queue Service (SQS) and Simple Notification Service (SNS). SQS can send, store, and receive messages between software components at any volume. SNS is similar, but it sends notifications to users based on a publish/subscribe (pub/sub) model. This means that users get associated with sns "topics," and by sending a notification to that topic, it will get shotgunned out to all of its subscribers.

## Additional Compute Services
EC2 is flexible, reliable, and scalable, but you're responsible for a lot of the configuration and achitecting of your applications. There are also "serverless" options offered by AWS, which means you can't see or access the underlying instances.

AWS Lambda is one serverless option. You deploy your code to an instance that is maintained and handled for you, and you don't really have to worry about the nuts and bolts of how things get run. Note that aws lambda is intended for brief work, on the order of 15 minutes, rather than long, drawn out processing.

If you need more control than that, you can look into Amazon Elastic Container Service (ECS). Or Amazon Elastic Kubernetes Service (EKS). These are both docker/kubernetes container orchestration platforms, and they can help you manage applications distributed across multiple EC2 instances. But if you want to avoid the effort of EC2 management, you can also look at AWS Fargate, which is a serverless compute platform for ECS and EKS.
